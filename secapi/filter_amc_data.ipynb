Script ยิง API เก็บ risk_spectrum และ spec_code ของแต่ละ กองทุนภายใน บลจ. ที่เก็บไว้ตอนแรก

```python
import time
import csv
header = ['proj_id','risk_spectrum','spec_code']
for index, row in df.iterrows():
    list_risk = []
    response = requests.get(f'https://api.sec.or.th/FundFactsheet/fund/amc/{row["unique_id"]}', headers = headers2)
    obj_dict = json.loads(response.text)
    filename = f'spec_suit/'+row["unique_id"]+'.csv'
    with open(filename, 'w',newline = '') as f_out:    
        f_out = csv.writer(f_out,quoting=csv.QUOTE_ALL)
        f_out.writerow(header)
        for x in obj_dict:
                suit_tmp = requests.get(f'https://api.sec.or.th/FundFactsheet/fund/{x["proj_id"]}/suitability', headers=headers2)
                spec_tmp = requests.get(f'https://api.sec.or.th/FundFactsheet/fund/{x["proj_id"]}/specification', headers=headers2)
                time.sleep(0.1)
                tmp_data = {}
                suit_json = {}
                spec_json = {}
                try:
                    # suit
                    suit_json = json.loads(suit_tmp.text)
                    tmp_data['proj_id'] = x['proj_id']
                    tmp_data['risk_spectrum'] = suit_json['risk_spectrum']
                except Exception as e1:
                    print(f"Error in Suit: {e1}")

                try:
                    # spec
                    spec_json = json.loads(spec_tmp.text)
                    tmp_data['proj_id'] = x['proj_id']
                    tmp_data['spec_code'] = ','.join(y['spec_code'] for y in spec_json)
                except Exception as e2:
                    print(f"Error in Spec: {e2}")
                if not tmp_data == {}:
                    f_out.writerow([tmp_data.get('proj_id'), tmp_data.get('risk_spectrum'), tmp_data.get('spec_code')])
                    
```

Scan files from pdf_factsheet folder to find all the files that have .txt extension

```python
import os

folder_path = 'pdf_factsheet'
target_extension = '.txt'  
not_delete_list = []
file_names = [f for f in os.listdir(folder_path) if f.endswith(target_extension) and os.path.isfile(os.path.join(folder_path, f))]

for file_name in file_names:
    file_name = file_name.replace('.txt','')
    not_delete_list.append(file_name)
```

Delete all pdf files that wasn't on the list (from the io Exception above the txt files from specific pdf files cannot be extract by pdfbox)

```python
import os

folder_path = 'pdf_factsheet'
all_files = os.listdir(folder_path)
for file_name in all_files:
    file_base_name, file_extension = os.path.splitext(file_name)
    if file_base_name not in not_delete_list and os.path.isfile(os.path.join(folder_path, file_name)):
        full_file_path = os.path.join(folder_path, file_name)
        try:
            os.remove(full_file_path)
            print(f"File {full_file_path} deleted successfully.")
        except FileNotFoundError:
            print(f"File {full_file_path} not found.")
        except Exception as e:
            print(f"Error deleting file {full_file_path}: {e}")
```

Scan all the extracted text files if it contains ": เสี่ยง" to find the according risk_spectrum and inplace the found value on 'risk_spectrum' column

```python
import re
for index, row in SCB_amc_rmf_ssf_filter_w_risk_spectrum.iterrows():
    with open(f'pdf_factsheet/{row["proj_id"]}.txt', 'r',  encoding="utf8") as file:
        data = file.read()
        match = re.search(r'(\S+)\s*:\s*เสี่ยง', data)
        if match:
            number_before_string = match.group(1)
            SCB_amc_rmf_ssf_filter_w_risk_spectrum.at[index, 'risk_spectrum'] = number_before_string
        else:
            print("Pattern not found on "+ row["proj_id"])
SCB_amc_rmf_ssf_filter_w_risk_spectrum
```
